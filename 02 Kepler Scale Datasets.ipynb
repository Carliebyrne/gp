# Scaling Gaussian Processes to Kepler-Sized Datasets

In the previous section, we demonstrated the use of a simple Gaussian process noise model. Unfortunately, any naïve implementation of these models will not scale to datasets as big as Kepler light curves. In this notebook, we benchmark a LAPACK-based likelihood computation against the implmentation in [George](https://github.com/dfm/george).

The key thing to remember here is that in astronomy, we normally don't know the noise model exactly so we generally want to marginalize over the hyperparameters of our kernel function. This means that the full solve + determinant must be computed millions of times for a single object.

```python
%matplotlib inline
from matplotlib import rcParams
rcParams["savefig.dpi"] = 150
```

```python
import kplr  # http://dan.iel.fm/kplr
import timeit
import numpy as np
import matplotlib.pyplot as pl
from scipy.linalg import cho_factor, cho_solve
```

```python
import george  # https://github.com/dfm/george
from george.kernels import RBFKernel
```

## Benchmark with Synthetic Data

We'll start by sampling some synthetic data from a Gaussian process to use for benchmarking. The parameters that I've chosen are designed to be similar to 2 months (a full quarter was too long for computational reasons) of long cadence Kepler data. 

```python
x = np.arange(0, 60., 0.5/24.)
yerr = 5e-5 + 1e-7 * np.random.randn(len(x))

a, s = 1e-8, 0.9
y = george.GaussianProcess(a*RBFKernel(s)).sample_prior(x) + yerr * np.random.randn(len(yerr))
```

```python
pl.plot(x, y, ".k", alpha=0.3);
```

Now we'll define the 2 likelihood functions that we want to benchmark against each other. One uses the scipy implementation of the Cholesky decomposition and the other uses [George](https://github.com/dfm/george).

```python
def simple_gp_lnlike(x, y, yerr, lna, lns):
    a, s = np.exp(lna), np.exp(lns)
    C = np.diag(yerr**2) + a * np.exp(-0.5 * (x[:, None] - x[None, :])**2 / (s*s))
    factor, flag = cho_factor(C)
    logdet = np.sum(2*np.log(np.diag(factor)))
    return -0.5 * (np.dot(y, cho_solve((factor, flag), y)) + logdet + len(x)*np.log(2*np.pi))

def george_lnlike(x, y, yerr, lna, lns):
    a, s = np.exp(lna), np.exp(lns)
    kernel = a * RBFKernel(s)
    gp = george.GaussianProcess(kernel, tol=1e-12, nleaf=100)
    gp.compute(x, yerr)
    return gp.lnlikelihood(y)
```

```python
print(george_lnlike(x[:100], y[:100], yerr[:100], np.log(a), np.log(s))
      -simple_gp_lnlike(x[:100], y[:100], yerr[:100], np.log(a), np.log(s)))
```

```python
ns = np.array([100, 200, 500, 1000, 1500, 2000, len(x)], dtype=int)
george_ts = np.empty(len(ns))
scipy_ts = np.empty(len(ns))
for i, n in enumerate(ns):
    george_ts[i] = timeit.timeit("george_lnlike(x[:{0}], y[:{0}], yerr[:{0}], {1}, {2})"
                                 .format(n, np.log(a), np.log(s)),
                                 setup="from __main__ import george_lnlike, x, y, yerr",
                                 number=3)
    scipy_ts[i] = timeit.timeit("simple_gp_lnlike(x[:{0}], y[:{0}], yerr[:{0}], {1}, {2})"
                                 .format(n, np.log(a), np.log(s)),
                                 setup="from __main__ import simple_gp_lnlike, x, y, yerr",
                                 number=3)
```

```python
pl.plot(np.log10(ns), np.log10(george_ts), "-k")
pl.plot(np.log10(ns), np.log10(george_ts), ".k")
pl.plot(np.log10(ns), np.log10(ns * np.log(ns) * george_ts[2] / (ns[2]*np.log(ns[2]))), "--k")
pl.plot(np.log10(ns), np.log10(scipy_ts), "-r")
pl.plot(np.log10(ns), np.log10(scipy_ts), ".r")
pl.plot(np.log10(ns), np.log10(ns ** 3 * scipy_ts[2] / (ns[2]**3)), "--r")
pl.ylim(-3, 2)
pl.ylabel(r"$\log_{10} \mathrm{runtime} / \mathrm{seconds}$")
pl.xlabel(r"$\log_{10} N$")
pl.savefig("figures/scaling-synthetic.pdf");
```

## Downloading Light Curves from MAST

We'll start by downloading a light curve from MAST. We'll get both the short and long cadence data for KOI-1474 in quarter 5 (actually only the last month of quarter 5 where we have both cadences). This is pretty easy if we use [kplr](http://dan.iel.fm/kplr):

```python
client = kplr.API()
lc = client.light_curves(12365184, ktc_target_type="LC", sci_data_quarter=5)[0]
with lc.open() as hdus:
    hdu = hdus[1]
    x_lc, y_lc, yerr_lc, q_lc = hdu.data["TIME"], hdu.data["SAP_FLUX"], hdu.data["SAP_FLUX_ERR"], hdu.data["SAP_QUALITY"]

sc = client.light_curves(12365184, ktc_target_type="SC", sci_data_quarter=5)[0]
with sc.open() as hdus:
    hdu = hdus[1]
    x_sc, y_sc, yerr_sc, q_sc = hdu.data["TIME"], hdu.data["SAP_FLUX"], hdu.data["SAP_FLUX_ERR"], hdu.data["SAP_QUALITY"]
```

Mask missing data and pull out only the last month where there is both long and short cadence data but no transits.

```python
m = np.isfinite(x_lc) * np.isfinite(y_lc) * np.isfinite(yerr_lc) * (x_lc > 504) * (q_lc == 0)
x_lc, y_lc, yerr_lc = x_lc[m], y_lc[m], yerr_lc[m]
m = np.isfinite(x_sc) * np.isfinite(y_sc) * np.isfinite(yerr_sc) * (x_sc > 504) * (q_sc == 0)
x_sc, y_sc, yerr_sc = x_sc[m], y_sc[m], yerr_sc[m]
```

Normalize the data.

```python
mu = np.median(y_lc)
y_lc, yerr_lc = y_lc / mu - 1, yerr_lc / mu
mu = np.median(y_sc)
y_sc, yerr_sc = y_sc / mu - 1, yerr_sc / mu
```
