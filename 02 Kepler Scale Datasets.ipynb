# Scaling Gaussian Processes to Kepler-Sized Datasets

In the previous section, we demonstrated the use of a simple Gaussian process noise model. Unfortunately, any naïve implementation of these models will not scale to datasets as big as Kepler light curves.

```python
%matplotlib inline
from matplotlib import rcParams
rcParams["savefig.dpi"] = 150
```

```python
import kplr  # http://dan.iel.fm/kplr
import numpy as np
import matplotlib.pyplot as pl
```

```python
import george  # https://github.com/dfm/george
from george.kernels import RBFKernel
```

We'll start by defining our two log-likelihood functions.

```python
def simple_gp_lnlike(x, y, yerr, lna, lns):
    a, s = np.exp(lna), np.exp(lns)
    C = np.diag(yerr**2) + a * np.exp(-0.5 * (x[:, None] - x[None, :])**2 / s)
    s, logdet = np.linalg.slogdet(C)
    if s <= 0:
        return -np.inf
    return -0.5 * (np.dot(y, np.linalg.solve(C, y)) + logdet + len(x)*np.log(2*np.pi))

def george_lnlike(x, y, yerr, lna, lns):
    a, sqrts = np.exp(lna), np.exp(0.5 * lns)
    kernel = a * RBFKernel(sqrts)
    gp = george.GaussianProcess(kernel, tol=1e-16, nleaf=20)
    gp.compute(x, yerr)
    return gp.lnlikelihood(y)
```

Load some long and short cadence data for the same quarter:

```python
client = kplr.API()
kic = client.star(12365184)
lc = client.light_curves(12365184, ktc_target_type="LC", sci_data_quarter=5)[0]
sc = client.light_curves(12365184, ktc_target_type="SC", sci_data_quarter=5)[0]
```

```python
with lc.open() as hdus:
    hdu = hdus[1]
    x_lc, y_lc, yerr_lc, q_lc = hdu.data["TIME"], hdu.data["SAP_FLUX"], hdu.data["SAP_FLUX_ERR"], hdu.data["SAP_QUALITY"]

with sc.open() as hdus:
    hdu = hdus[1]
    x_sc, y_sc, yerr_sc, q_sc = hdu.data["TIME"], hdu.data["SAP_FLUX"], hdu.data["SAP_FLUX_ERR"], hdu.data["SAP_QUALITY"]
```

Mask missing data and pull out only the last month where there is both long and short cadence data.

```python
m = np.isfinite(x_lc) * np.isfinite(y_lc) * np.isfinite(yerr_lc) * (x_lc > 504) * (q_lc == 0)
x_lc, y_lc, yerr_lc = x_lc[m], y_lc[m], yerr_lc[m]
m = np.isfinite(x_sc) * np.isfinite(y_sc) * np.isfinite(yerr_sc) * (x_sc > 504) * (q_sc == 0)
x_sc, y_sc, yerr_sc = x_sc[m], y_sc[m], yerr_sc[m]
```

Normalize the data.

```python
mu = np.median(y_lc)
y_lc, yerr_lc = y_lc / mu - 1, yerr_lc
mu = np.median(y_sc)
y_sc, yerr_sc = y_sc / mu - 1, yerr_sc
```

```python
pl.plot(x_sc, y_sc, ".k", alpha=0.5);
pl.plot(x_lc, y_lc, ".r", alpha=0.5);
```
