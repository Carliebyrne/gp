## Search

The key speed advantage in the search phase is that we can pre-factorize the covariance matrix and then evaluate many models quickly:

```python
a, s = truth[:2]
gp = george.GaussianProcess(a * RBFKernel(s), tol=1e-16, nleaf=20)
gp.compute(t, fe)
```

Then we loop over a grid of times and compute the delta-log-likelihood for a box model versus the null at each phase:

```python
def box_model(t, t0, duration, depth):
    m = np.ones_like(t)
    m[np.abs(t-t0) < 0.5*duration] -= depth
    return m
```

```python
times = np.linspace(min(t), max(t), 1500)
gp_lls = np.empty_like(times)
mf_lls = np.empty_like(times)
print("Running {0} hypothesis tests".format(len(times)))
for i, t0 in enumerate(times):
    mod = box_model(t, t0, 0.4, 0.019**2)
    gp_lls[i] = gp.lnlikelihood(f - mod)
    mf_lls[i] = -0.5 * np.sum(((detrend-mod)/fe)**2)
print("Finished")
```

```python
gp_null = gp.lnlikelihood(f - 1)
mf_null = -0.5 * np.sum(((detrend-1)/fe)**2)
```

```python
pl.gca().axhline(0, color="k")
pl.plot(times, mf_lls-mf_null, "k", lw=2)
pl.plot(times, gp_lls-gp_null, "r")
pl.xlabel("time [KBJD]")
pl.ylabel(r"$\Delta \ell$")
pl.xlim(min(t), max(t))
pl.savefig("figures/injection-dll.pdf");
```

## Stellar Variability

```python
client = kplr.API()
lc = client.light_curves(3223000, sci_data_quarter=7)[0]
```

```python
data = lc.read()
t, f, fe, q = data["TIME"], data["SAP_FLUX"], data["SAP_FLUX_ERR"], data["SAP_QUALITY"]
m = np.isfinite(t) * np.isfinite(f) * np.isfinite(fe) * (t > 635) * (t < 660) * (q==0)
t, f, fe = t[m], f[m], fe[m]
mu = np.median(f)
f, fe = f / mu - 1, fe / mu
```

```python
pl.plot(t, f, ".k")
pl.xlabel("time [KBJD]")
pl.ylabel("relative flux")
pl.savefig("figures/kic3223000.pdf");
```

```python
periods = np.linspace(1, 10, 40)
p_lls = np.zeros_like(periods)
for i, per in enumerate(periods):
    kernel = 1e-9 * Matern32Kernel(10.0) * CosineKernel(per) + 1e-4 * RBFKernel(1.0)
    gp = george.GaussianProcess(kernel)
    gp.compute(t, fe)
    p_lls[i] = gp.lnlikelihood(f)
```

```python
pl.plot(periods, p_lls/1e6, "k", lw=2)
pl.ylabel(r"log likelihood [$\times 10^6$]")
pl.xlabel("rotation period")
pl.title("KIC 3223000");
```
