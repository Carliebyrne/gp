```python
import emcee
import numpy as np
import matplotlib.pyplot as pl
from IPython.display import display, Math, Latex
```

```python
np.random.seed(123456)  # Reproducible!!
```

## Generating a Synthetic Dataset

First, build the "true" dataset with $N=50$ datapoints from a line model $y = m\,x+b$.

```python
true_m, true_b = 0.5, -0.25
N = 50
x = 10 * np.sort(np.random.rand(N)) - 5
y = true_m * x + true_b
```

Introduce some noise with both measurement uncertainties and non-trivial correlated errors.

```python
yerr = 0.1 + 0.4 * np.random.rand(N)
true_cov = 0.5 * np.exp(-0.5 * (x[:, None]-x[None, :])**2 / 1.3**2) + np.diag(yerr ** 2)
y = np.random.multivariate_normal(y, true_cov)
```

Let's plot the structure of the covariance matrix that generated the data:

```python
pl.imshow(true_cov, interpolation="nearest", cmap="gray");
```

And plot the data with the observational uncertainties. The true line is plotted in black.

```python
x0 = np.linspace(-6, 6, 5000)
pl.errorbar(x, y, yerr=yerr, fmt=".k", capsize=0)
pl.plot(x0, true_m * x0 + true_b, "k", lw=2, alpha=0.8);
```

## Assuming Uncorrelated Noise

If we ignore the correlated uncertainties and only use the "observational errors", the maximum likelihood line is given by:

$$\left [ \begin{array}{c} m \\ b \end{array} \right ] = \mathbf{S} \, \mathbf{A}^\mathrm{T} \, \mathbf{C}^{-1} \, \mathbf{y}$$

where

$$\mathbf{A} = \left [ \begin{array}{cc} x_1 & 1 \\ x_2 & 1 \\ \vdots & \vdots \\ x_n & 1 \end{array} \right ] \quad,\quad \mathbf{C} = \left [ \begin{array}{cccc} \sigma_1^2 &0&\cdots&0 \\ 0& \sigma_2^2 & \cdots & 0 \\ \vdots & \vdots & \ddots & \vdots \\ 0 & 0 & \cdots & \sigma_n^2 \end{array} \right ] \quad,\quad \mathbf{y} = \left [ \begin{array}{c} y_1 \\ y_2 \\ \vdots \\ y_n \end{array} \right ] \quad, \quad\mathrm{and} \quad \mathbf{S} = \left [ \mathbf{A}^\mathrm{T}\,\mathbf{C}^{-1}\,\mathbf{A} \right ]^{-1}\quad.$$

The $\mathbf{S}$ matrix is (in this very specific case) the covariance matrix for the posterior constraints on $m$ and $b$.

Let's compute this for our case:

```python
A = np.vander(x, 2)
C = np.diag(yerr**2)
Sinv = np.dot(A.T, np.linalg.solve(C, A))
ls_m, ls_b = np.linalg.solve(Sinv, np.dot(A.T, np.linalg.solve(C, y)))
ls_S = np.linalg.inv(Sinv)
```

```python
display(Math(r"m_\mathrm{{LS}} = {0:.3f} \pm {1:.3f}".format(ls_m, np.sqrt(ls_S[0, 0]))))
display(Math(r"b_\mathrm{{LS}} = {0:.3f} \pm {1:.3f}".format(ls_b, np.sqrt(ls_S[1, 1]))))
```

Here, we'll compute the posterior constraints calculated under this assumption of independent uncertainties and plot it on top of the data:

```python
samples = np.dot(np.vander(x0, 2), np.random.multivariate_normal([ls_m, ls_b], ls_S, size=5000).T)
ls_mu = np.mean(samples, axis=1)
ls_std = np.std(samples, axis=1)
```

```python
pl.errorbar(x, y, yerr=yerr, fmt=".k", capsize=0)
pl.plot(x0, true_m * x0 + true_b, "k", lw=2, alpha=0.8);
pl.fill_between(x0, ls_mu+ls_std, ls_mu-ls_std, color="r", alpha=0.3);
```

This is obviously an overly optimistic result! The red area represents the "1$\sigma$" constraint. Not only are the results wrong but the uncertainties also don't include the "truth".

## Using the True Covariance Structure

If we know—or have an estimate of—the covariances between observations, the problem is still Gaussian so the solution is still closed form. The math is the same as above but the covariance matrix $\mathbf{C}$ is dense instead of diagonal. We saved the true covariance structure above in the matrix `true_cov` so let's use it and compute the posterior constraints on $m$ and $b$.

```python
Sinv = np.dot(A.T, np.linalg.solve(true_cov, A))
corr_m, corr_b = np.linalg.solve(Sinv, np.dot(A.T, np.linalg.solve(true_cov, y)))
corr_S = np.linalg.inv(Sinv)
```

```python
display(Math(r"m_\mathrm{{corr}} = {0:.3f} \pm {1:.3f}".format(corr_m, np.sqrt(corr_S[0, 0]))))
display(Math(r"b_\mathrm{{corr}} = {0:.3f} \pm {1:.3f}".format(corr_b, np.sqrt(corr_S[1, 1]))))
```

This looks better! Let's plot the constraints in the data plane and compare to the previous result:

```python
samples = np.dot(np.vander(x0, 2), np.random.multivariate_normal([corr_m, corr_b], corr_S, size=5000).T)
corr_mu = np.mean(samples, axis=1)
corr_std = np.std(samples, axis=1)
```

```python
pl.errorbar(x, y, yerr=yerr, fmt=".k", capsize=0)
pl.plot(x0, true_m * x0 + true_b, "k", lw=2, alpha=0.8);
pl.fill_between(x0, corr_mu+corr_std, corr_mu-corr_std, color="b", alpha=0.3)
pl.fill_between(x0, ls_mu+ls_std, ls_mu-ls_std, color="r", alpha=0.3);
```

Boom. As expected, our fit is now consistent with the truth. Are we done? Never.

## Unknown Covariances

In The Real World™, we rarely actually know the covariance structure of our data. Instead, we need to build a model of it. In this case, let's model the noise using the following parametric form (which, in this case, just happens to include the truth):

$$\mathbf{C} = \left [ C_{ij} \right ] \quad \mathrm{where} \quad C_{ij} = \sigma_i^2\,\delta_{ij} + k(x_i, x_j) \quad \mathrm{and} \quad k(x_i, x_j) = a \, \exp \left(-\frac{(x_i-x_j)^2}{2\,s} \right ) \quad.$$

In this equation, $a$ and $s$ are now parameters that we want to fit for (or marginalize over). As a result, the solution for $m$ and $b$ only has a closed form when $a$ and $s$ are *fixed*. To do the full posterior sampling (marginalization), we'll use MCMC but we need to write down the likelihood function. You probably remember (or can [look up on Wikipedia](http://en.wikipedia.org/wiki/Multivariate_normal_distribution)) the properly normalized PDF for a multivariate Gaussian and it looks something like this:

$$\log p(\mathbf{y}\,|\,m,\,b,\,a,\,s) = -\frac{1}{2}\,\mathbf{r}^\mathrm{T}\,\mathbf{C}^{-1}\,\mathbf{r} - \frac{1}{2}\log \det\mathbf{C} - \frac{N}{2} \log 2\,\pi$$

where $N$ is the number of data points and

$$\mathbf{r} = \left [ \begin{array}{ccc} y_1 - (m\,x_1+b) & \cdots & y_n - (m\,x_n+b) \end{array} \right ]^\mathrm{T} \quad.$$

The important thing to remember here is that the first two terms on the right hand side depend on the model parameters through the construction of the $\mathbf{C}$ matrix. In code, this likelihood function (ignoring the constant last term) would looks something like:

```python
def lnlike(m, b, lna, lns):
    a, s = np.exp(lna), np.exp(lns)
    C = np.diag(yerr**2) + a * np.exp(-0.5 * (x[:, None] - x[None, :])**2 / s)
    s, logdet = np.linalg.slogdet(C)
    if s <= 0:
        return -np.inf
    r = y - (m*x + b)
    return -0.5 * (np.dot(r, np.linalg.solve(C, r)) + logdet)
```

If we want to use MCMC, we need to compute the posterior probability up to a constant. To do this, we'll assume uniform priors on $m$, $b$, $\ln a$, and $\ln s$ and apply Bayes' theorem:

```python
def lnprior(m, b, lna, lns):
    if not (-2 < m < 2 and -2 < b < 2 and -5 < lna < 5 and -5 < lns < 5):
        return -np.inf
    return 0.0

def lnprob(p):
    lp = lnprior(*p)
    if not np.isfinite(lp):
        return -np.inf
    return lp + lnlike(*p)
```

Now that we have this probability distribution, we'll use [emcee](http://dan.iel.fm/emcee/) to sample from it. Take a look at [the documentation](http://dan.iel.fm/emcee/) for more information on how to use it but for our case, it is quite simple.

First, we initialize the "walkers" cleverly:

```python
ndim, nwalkers = 4, 32
p0 = np.array([true_m, true_b, np.log(0.5), np.log(1.3)])
pos = [p0 + 1e-2 * np.random.randn(ndim) for i in range(nwalkers)]
```

And then set up the sampler and run a burn-in:

```python
sampler = emcee.EnsembleSampler(nwalkers, ndim, lnprob)
pos, lp, state = sampler.run_mcmc(pos, 300)
```

Rest the sampler and run a production run:

```python
sampler.reset()
pos, lp, state = sampler.run_mcmc(pos, 1000)
chain = sampler.chain
```

We'll look at the traces to get an idea of how the sampling went.

```python
fig, axes = pl.subplots(4, 1, figsize=(5, 6), sharex=True)
fig.subplots_adjust(left=0.1, bottom=0.1, right=0.96, top=0.98,
                    wspace=0.0, hspace=0.05)
[a.plot(np.arange(chain.shape[1]), chain[:, :, i].T, "k", alpha=0.5)
 for i, a in enumerate(axes)]
[a.set_ylabel("${0}$".format(l)) for a, l in zip(axes, ["m", "b", "\ln a", "\ln s"])]
axes[-1].set_xlim(0, chain.shape[1])
axes[-1].set_xlabel("iteration");
```

And finally, plot the posterior constraints and compare them to the constraints obtained when the covariances were known:

```python
samples = np.dot(np.vander(x0, 2), chain[:, ::21, :2].reshape((-1, 2)).T)
mcmc_mu = np.mean(samples, axis=1)
mcmc_std = np.std(samples, axis=1)
```

```python
pl.errorbar(x, y, yerr=yerr, fmt=".k", capsize=0)
pl.plot(x0, true_m * x0 + true_b, "k", lw=2, alpha=0.8);
pl.fill_between(x0, mcmc_mu+mcmc_std, mcmc_mu-mcmc_std, color="g", alpha=0.3)
pl.fill_between(x0, corr_mu+corr_std, corr_mu-corr_std, color="b", alpha=0.3)
pl.fill_between(x0, ls_mu+ls_std, ls_mu-ls_std, color="r", alpha=0.3);
```

The colors are the same as before but the green area indicates the marginalized posterior constraint.
